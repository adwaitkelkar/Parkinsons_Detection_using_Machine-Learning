```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn import svm
from sklearn.metrics import accuracy_score
```

```python
df=pd.read_csv("/content/sample_data/parkinsons.csv")
```

```python
#Grouping Data Based on Target variable(taking mean)
df.groupby('status').mean()
```

```python
df=df.drop('name',1)
```

```python
X=df.drop(columns=['status'],axis=1)
Y=df['status']
```

```python
X
```

```python
Y
```

```python
#Spliting Data Set
X_train, X_test, Y_train, Y_test=train_test_split(X,Y,test_size=0.2,random_state=2)
```

```python
print(X.shape, X_test.shape, X_train.shape)
```

```python
#Data Standardization
scaler=StandardScaler()
```

```python
scaler.fit(X_train)
```

```python
X_train= scaler.transform(X_train)
```

```python
X_test=scaler.transform(X_test)
```

```python
X_train
```

```python
#Model Traing with SVM
model=svm.SVC(kernel='linear')
```

```python
#traning SVM Model 
model.fit(X_train,Y_train)
```

```python
#model Evaluation 
#Accuracy Score on training data
X_train_predic=model.predict(X_train)
train_acc=accuracy_score(Y_train,X_train_predic)
```

```python
print("Accuracy of training Data",train_acc)
```

```python
#Accuracy Score on test data
X_test_predic=model.predict(X_test)
test_acc=accuracy_score(Y_test,X_test_predic)
```

```python
print("Accuracy of training Data",test_acc)
```

```python
#Since our Accuracy of Train and test data is almost same we can say that there is no sope of the model being overfitted
```
